## loc and iloc indexers

- iloc allows us to access rows by integer location and returns a Series that contains the values of the data. Usees integers only.

    - index becomes column names

    df.iloc[0]   returns the first row
    
    
- loc allows us to search by label for row when your index is custom with labels, otherwise it will still use the integers as labels if your index is still integers.

```python
df.iloc[0]
```

```python
# returns a DataFrame with the multiple rows

df.iloc[[0,1]]
```

```python
# select columns as the second argument to return the second column for the first two rows

df.iloc[[0,1], 1]
```

```python
# it looks like iloc right now because our index is still made up of integers
# all of the responses for the first respondent

df.loc[0]
```

```python
# The response to the Hobbyist column for the first respondent

df.loc[0, 'Hobbyist']
```

```python
# here we use integer labels from our index but we can use string labels to index the columns

df.loc[[0,1], ['Student','Hobbyist']]
```

```python
# responses to the Hobbyist column for the first four respondents. Indexing inclusive in .loc

df.loc[0:3, 'Hobbyist']
```

```python
# This is an exmple of WHY .loc is inclusive... because of the labels.

df.loc[0:3, 'Hobbyist': 'Employment']
```
