- Plot y = x^2

```python
x = range(-4, 5)
y = [x**2 for x in x]

plt.plot(x,y, marker='x')
plt.show()
```

```python
x = [-1, 0, 1, 2, 3, 4, 5]
y = [x**2 for x in x]

plt.plot(x, y, marker='o')
plt.show()
```

```python
df['class'].value_counts()
```

```python
city_mileage = df.cty
hwy_mileage = df.hwy
```

```python
plt.hist((city_mileage, hwy_mileage), bins=15, align='left', edgecolor='black', alpha=.7, label=['city', 'highway'])

plt.title('MPG by Type of Driving')
plt.ylabel('Count')
plt.xlabel('Type of Driving')
plt.legend()
plt.show()
```

```python
suv_df = df[df['class'] == 'suv']
```

```python
plt.figure(figsize=(12,10))
plt.subplot(4, 2, 1)
plt.hist(suv_df['cty'], bins=10, edgecolor='black')
```

```python
plt.figure(figsize=(16, 6))

plt.subplot(1, 2, 1)
plt.hist(city_mileage, bins=20, color='orange', edgecolor='black')
plt.xlabel('MPG')
plt.ylabel('Count')
plt.title('City Mileage')

plt.subplot(1, 2, 2)
plt.hist(hwy_mileage, bins=20, edgecolor='black')
plt.xlabel('MPG')
plt.ylabel('Count')
plt.title('Highway Mileage')

plt.suptitle('Mileage Distributions by Type')

plt.show()
```

```python
import math

x = range(10)
y = [math.sqrt(x) for x in x]
```

```python
plt.plot(y)
plt.title('y=square root x')
plt.annotate('Intersection', xy=(0, 0), xytext=(1,1.5),
             arrowprops={'facecolor': 'blue'})

plt.show()
```

```python
from pydataset import data
```

```python
df = data('iris')
df.head(1)
```

```python
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width',
       'species']
```

```python
df.head(1)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))
fig.subplots_adjust(hspace=0.8)
fig.suptitle('Distributions of Iris Features')

for ax, feature, name in zip(axes.flatten(), df.drop('species',axis=1).values.T, df.columns.values):
    ax.hist(feature, bins=len(np.unique(df.drop('species',axis=1).values.T[0])//2))
    ax.set(title=name[:-4].upper(), xlabel='cm')
```
