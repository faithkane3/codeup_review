```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = list(range(150))

plt.plot(x)
plt.show()
```

- When given only one variable, it will be treated as y, and x will be the index of the Series.

```python
y = [1, 2, 3, 4, 5]
```

```python
plt.plot(y)
plt.show()
```

- Plot y = x^2

```python
x = range(-4, 5)
y = [x**2 for x in x]

plt.plot(x,y, marker='x')
plt.show()
```

```python
x = [-1, 0, 1, 2, 3, 4, 5]
y = [x**2 for x in x]

plt.plot(x, y, marker='o')
plt.show()
```

```python
from random import randint

# Some random data
x1 = [randint(-5, 25) + n for n in range(150)]
x2 = [x + randint(-6, 6) for x in x1]

# we can use the alpha kwarg to define how transparent the color is
plt.plot(x1, c='green', alpha=0.6)
plt.plot(x2, c='red', alpha=0.4)
plt.show()
```

```python
from pydataset import data
```

```python
data('mpg', show_doc=True)
```

```python
df = data('mpg')
```

```python
df.head()
```

```python
city_mileage = df.cty
hwy_mileage = df.hwy
```

```python
plt.hist(city_mileage, bins=20, color='orange', edgecolor='black')
plt.xlabel('MPG')
plt.ylabel('Count')
plt.title('City Mileage')

plt.show()
```

```python
plt.hist(hwy_mileage, bins=20, edgecolor='black')
plt.xlabel('MPG')
plt.ylabel('Count')
plt.title('Highway Mileage')

plt.show()
```
